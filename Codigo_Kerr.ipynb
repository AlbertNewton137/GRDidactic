```python
import sympy as smp
import time


def Christoffel(i,j,k):
    s = 0
    for a in range(n):
        s+=1/2*g_inv[a,i]*(smp.diff(g[a,j],coord[k])+smp.diff(g[a,k],coord[j])-smp.diff(g[j,k],coord[a]))
    return smp.simplify(s)

def Riemann(i,j,k,l):
    p=smp.diff(ch[i,j,l],coord[k])-(smp.diff(ch[i,j,k],coord[l]))
    s=0
    for o in range(n):
        s+=ch[i,o,k]*ch[o,j,l]-(ch[i,o,l]*ch[o,j,k])
    return smp.simplify(p+s)

def Ricci(i,j):
    s=0
    for o in range(n):
        s+=riem[o,i,o,j]
    return smp.simplify(s)

def Einstein(i,j):
    s=-1/2*g[i,j]*er+ricci[i,j]
    return smp.simplify(s)
```

```python
coord = smp.symbols('t r theta phi', positive = True)

rs,A,C,= smp.symbols('r_s a c', positive = True)
sigma = smp. symbols('Sigma',cls=smp.Function, positive = True)
sigma = sigma(coord[1],coord[2])
delta = smp. symbols('Delta',cls=smp.Function, positive = True)
delta = delta(coord[1])

valor_sigma = coord[1]**2+A**2*smp.cos(coord[2])**2
valor_delta = coord[1]**2-rs*coord[1]+A**2

n = len(coord)

g= smp.Matrix([[0 for i in range(n)] for i in range(n)])
ch=  smp.MutableDenseNDimArray([[[0 for i in range(n)] for i in range(n)] for i in range(n)])
riem = smp.MutableDenseNDimArray([[[[0 for i in range(n)] for i in range(n)] for i in range(n)]for i in range(n)])
ricci = smp.Matrix([[0 for i in range(n)] for i in range(n)])
einstein = smp.Matrix([[0 for i in range(n)] for i in range(n)])

g[0,0] = -(1-rs*coord[1]/sigma)*C**2
g[0,3] = -rs*A*coord[1]*smp.sin(coord[2])**2/sigma*C
g[3,0] = -rs*A*coord[1]*smp.sin(coord[2])**2/sigma*C
g[1,1] = sigma/delta
g[2,2] = sigma
g[3,3] = (coord[1]**2+A**2+rs*A**2*coord[1]*smp.sin(coord[2])**2/(sigma))*smp.sin(coord[2])**2



g_inv = smp.simplify(g.inv())


gcovariante = smp.symbols(fr'g_alpha_beta')
gcontravariante = smp.symbols(fr'g^alpha^beta')

display(smp.Eq(gcovariante,g,evaluate=False))
display(smp.Eq(gcontravariante,g_inv,evaluate=False))
```

```python
#Susutitucion
subs = {sigma: valor_sigma,
         delta: valor_delta,
         smp.diff(sigma,coord[1]): smp.diff(valor_sigma,coord[1]),
         smp.diff(sigma,coord[2]): smp.diff(valor_sigma,coord[2]),
         smp.diff(delta,coord[1]): smp.diff(valor_delta,coord[1])}
```

```python
%%time
#Simbolos de Cristoffel
for a in range(n):
    for b in range(n):
        for c in range(n):
            ch[a,b,c]=Christoffel(a, b, c)
            if ch[a, b, c] != 0:
                gamma = smp.Symbol(fr'\Gamma_{coord[b]}_{coord[c]}^{coord[a]}')
                ch[a,b,c]=smp.simplify(ch[a,b,c].subs(subs))
                display(smp.Eq(gamma,ch[a, b, c],evaluate=False))
```
